```python
# Write a Python program to get the Fibonacci series between 0 to 50.


def fib():
  for x in range(0,50):
    if x == 0:
      print(0)
    elif x==1:
      print(1)
    else:
      print((x-2)+(x-1))


```

```python
fib()
```
